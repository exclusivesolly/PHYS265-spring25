```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cmap
import matplotlib.cm as cm
from scipy.integrate import quad
from scipy import constants
```

```python
#Constants
G = 9.8                 #Graviational Acceleration    m/s^2
M_earth = 5.9e24        #Mass of the Earth            kg
M_moon = 7.3e22         #Mass of the Moon             kg
M_apollo = 5500         #Mass of Apollo               kg
d_earthtomoon = 3.8e8   #Distacne from Earth to Moon  km
wet_mass = 2.8e6        #Wet Mass of S-1C             kg
dry_mass = 7.5e5        #Dry Mass of S-1C             kg
burn_rate = 1.3e4       #Burn Rate of S-1C            kg/s
exhaust_v = 2.4e3       #Exhuast Velocity of S-1c     m/s
```

**PART 1: The Gravitational Potential of the Earth**

```python
def GP_e(M_earth, Xe, Ye, x, y): #GP_e is Gravitational Potential of the Earth
    if np.isclose(Xe, x) and np.isclose(Ye, y):
        r = np.nan
    else:
        r = np.sqrt((x - Xe)**2 + (y - Ye)**2)
    phi = -(constants.G*M_earth)/r
    return phi
```

```python
GP_e_v = np.vectorize(GP_e)
x = np.linspace(0, 576000000, 1000)  #1.5 * 3e8 = 576000000
y = 0
fig, ax = plt.subplots(1, 1)
ax.plot(x, np.abs(GP_e_v(M_earth, 0, 0, x, 0)))
ax.set_title("Absolute Value of Gravitational Potential")
ax.set_ylabel('Gravitational Potential (j)')
ax.set_xlabel("Distance from Earth's Surface (m)")
ax.set_yscale('log')
ax.grid(True)
```

```python
x_range = np.linspace(-576000000, 576000000, 1000)
y_range = np.linspace(-576000000, 576000000, 1000)
X, Y = np.meshgrid(x_range, y_range)
fig, ax = plt.subplots(1, 1)
mesh = ax.pcolormesh(X, Y, np.abs(GP_e_v(M_earth, 0, 0, X, Y)), norm = colors.LogNorm(vmin = 1e5, vmax = 1e7), cmap = cmap.jet)
ax.set_title("Gravitational Potential Energy with Earth at the Orgin")
ax.set_xlabel("X distance (m)")
ax.set_xlabel("Y distance (m)")
cbar = fig.colorbar(mesh)
cbar.set_label("Gravitational Potential Energy")
```

**PART 2: The Gravitational Potential of the Earth-Moon System**

```python
Xm = d_earthtomoon/np.sqrt(2)
Ym = d_earthtomoon/np.sqrt(2)
def GP_em(M_earth, M_moon, Xe, Ye, Xm, Ym, x, y):  #GP_em is Gravitational Potential of the Earth-Moon system
    if np.isclose(Xe, x) and np.isclose(Ye, y):
        Re = np.nan
    elif np.isclose(Xm, x) and np.isclose(Ym, y):
        Rm = np.nan
    else:
        Re = np.sqrt((x - Xe)**2 + (y - Ye)**2)
        Rm = np.sqrt((x - Xm)**2 + (y - Ym)**2)
        GP_em = -constants.G*(M_earth/Re + M_moon/Rm)
    return GP_em
GP_em_v = np.vectorize(GP_em)
```

```python
x1_range = np.linspace(-576000000, 576000000, 100)
y1_range = np.linspace(-576000000, 576000000, 100)
X1, Y1 = np.meshgrid(x1_range, y1_range)
fig, ax = plt.subplots(1, 1)
mesh = ax.pcolormesh(X1, Y1, np.abs(GP_em_v(M_earth, M_moon, 0, 0, Xm, Ym, X1, Y1)), norm = colors.LogNorm(vmin = 1e5, vmax = 1e7), cmap = cmap.jet)
ax.set_title("Gravitational Potential Energy with Earth-Moon System")
ax.set_xlabel("X distance (m)")
ax.set_xlabel("Y distance (m)")
cbar = fig.colorbar(mesh)
cbar.set_label("Gravitational Potential Energy")
```

```python
fig, ax = plt.subplots(1, 1)
level = np.linspace(1e5, 1e7, 100)
contour = ax.contour(X1, Y1, np.abs(GP_em_v(M_earth, M_moon, 0, 0, Xm, Ym, X1, Y1)), norm = colors.LogNorm(vmin = 1e5, vmax = 1e7), cmap = cmap.jet, levels = level)
ax.set_title("Gravitational Potential Energy with Earth-Moon System(contour)")
ax.set_xlabel("X distance (m)")
ax.set_xlabel("Y distance (m)")
cbar = fig.colorbar(contour)
cbar.set_label("Gravitational Potential Energy")
```

**PART 3: The Gravitational Force Field of the Earth-Moon System**

```python
def GF(M1, M2, x1, y1, x2, y2): #GF is Graviational Force
    if np.isclose(x1, x2) and np.isclose(y1, y2):
        fx = np.nan
        fy = np.nan
    else:
        r_squared = (x2 - x1)**2 + (y2 - y1)**2
        vec_r = np.array([x2 - x1, y2 - y1])
        vec_r_hat = vec_r/np.linalg.norm(vec_r)
    fx = (-constants.G*M1*M2/r_squared)*vec_r_hat[0]
    fy = (-constants.G*M1*M2/r_squared)*vec_r_hat[1]
    return fx, fy
```

```python
GF_v = np.vectorize(GF)

fig, ax = plt.subplots(1, 1)
Fex, Fey = GF_v(M_earth, M_apollo, 0, 0, X1, Y1)
Fmx, Fmy = GF_v(M_moon, M_apollo, Xm, Ym, X1, Y1)

Fx_t = Fex + Fmx
Fy_t = Fey + Fmy

c = np.log(np.hypot(Fx_t, Fy_t))
norm = colors.LogNorm(vmin = 0.6 , vmax = 6)
stream = ax.streamplot(X1, Y1, Fx_t, Fy_t, density = 2.5, color = c, norm = norm, cmap = cm.jet)
ax.set_title("Force exerted on the Apollo rocket by the Earth-Moon System")
ax.set_xlabel("X distance (m)")
ax.set_ylabel("Y distance (m)")
cbar = fig.colorbar(stream.lines)
cbar.set_label("Gravitational force of Earth-Moon System")
```

**PART 4: Altitude of the Saturn V Rocket**

```python
T = (wet_mass - dry_mass)/burn_rate

def V(T , M_wet, M_dry, burn_rate, Ex_s, G):   
    m = M_wet - burn_rate * T
    if m < M_dry:
        m = M_dry
    dv = Ex_s * np.log(M_wet/(m)) - G * T
    return dv

#Part C
from scipy.integrate import quad
h = quad(V, 0 , T, args =(wet_mass, dry_mass, burn_rate, exhaust_v, G))[0]


print(f"The height the rocket reaches at burnout is {h:.1f} meters")
print(f"The time before burnout is {T:.1f} seconds")
```
