# Jupyter Notebook cells

There are two types of cells you see here. There are the "Markdown" cells, which contain text. And the "Code" cells, which contain python code and can be executed. The active cell is the one with the (blue) bar on the left of the cell.

There are two ways to "execute" a cell.  You can SHIFT-ENTER or you can move the mouse to the black "play" triangle at the top of this window.

You can actually also execute a "Markdown" cell, which means the "Raw" text (in edit mode) in that cell is now displayed nicely. You will need to double click in a markdown cell to go into "Raw" mode to edit it.  Go and that that in this call.

Also note when you execute a cell, it automatically advances to the next cell.
