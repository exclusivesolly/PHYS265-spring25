# Notes

1) We can numerically estimate a definite integral by summing up a bunch of rectangular boxes inscribed inside the function. Each box has width dx, and the height
of each is the function value at each x location along the way. If we take the limit
as  $dx \rightarrow 0$, we get the Riemann definition of the definite integral.
Example: we can easily compute an estimate of the definite integrate of a quadratic
function $f(x) = x^2$ from 0 to 1. Here we use the .sum() method provided by numpy
for summing up arrays.

```
xx = np.linspace(0.0,0.99,100)      # dx = 0.01
yy = xx**2
print(yy.sum()*0.01)                # this is the Riemann sum for dx = 0.01
xx = np.linspace(0.0,0.999,1000)    # Now try dx = 0.001
print(yy.sum()*0.001)               # Riemann sum for dx = 0.001
```

2) To compute an indefinite integral, we let the upper limit of the integral be a
variable:
$$
F(x) = \int_{start}^{x} f(x')dx'   ~~~~\Leftrightarrow~~~~~   {{\partial F} \over {\partial x}} = f(x)
$$
where start is any x reference point we may choose (perhaps x = 0, for example).
We can compute F (x) numerically with quad() by calling it within a loop, where
the upper limit of quad’s integration steps along in x:

```
def f(x): return x**2            # let’s integrate this function
xx = np.linspace(0.0,1.0,101)    # x domain
F = np.zeros(len(xx))            # create space to store results
for i in range(len(xx)):
   F[i], err = quad(f, 0, x[i])  # integrate from zero to x[i]
plt.plot(xx,F)
```

3) The python **lambda function** is an alternative grammar for creating user defined
functions. It does not do anything beyond what the familiar `def` statement can
do, but `lambda` may make the code slightly less cluttered and easier to read.
Python coders tend to use lambda in situations where a function is needed once
and only once, and where the function definition is rather simple and compact,
and must be limited to one line.
Note: because lambda is a reserved word in python, we cannot use it to represent
a variable or value.
Thus the following two function definitions are identical:

```
def f(x, a=1, b=0):
    return x**2 + a*x + b

f = lambda x, a=1, b=0: x**2 + a*x + b
```


# Excersizes

Introducing a new `quad` import today:


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
````

````python
#    Showing integration steps (run this as a Code cell)

#          function to integrate
def f(x):
    return 3*x**2

#   or use:   f = lambda x: 3*x**2

xx = np.linspace(0,2,5)       # step function to show riemann steps
yy = f(xx)
xs = np.linspace(0,2,101)     # smooth function to show underlying function
ys = f(xs)

d0=0    # offsets between the 3 plots
d1=5
d2=10

fig, ax = plt.subplots()
ax.plot(xs,ys+d0, label='post')                # smooth function
ax.plot(xs,ys+d1, label=f'mid +{d1}')
ax.plot(xs,ys+d2, label=f'pre +{d2}')

ax.plot(xx,yy,'or')                # step function with red dots
ax.plot(xx,yy+d1,'or')
ax.plot(xx,yy+d2,'or')

                              # step function using a plot drawstyle
ax.plot(xx,yy+d0,drawstyle='steps-post', color='black')
ax.plot(xx,yy+d1,drawstyle='steps-mid',  color='black')
ax.plot(xx,yy+d2,drawstyle='steps-pre',  color='black')

ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Rieman integration and midpoint integrations for $y=3x^2$")
fig.tight_layout()
````

Integrate a simple quadratic function from 0 to 1 by summing, a.k.a. the
[Riemann integral](https://en.wikipedia.org/wiki/Riemann_integral), using manual looping.
 
Try dx = 0.01, then dx = 0.001. Perhaps try the trapezoid rule also.

````python
def f(x):
    return 3*x**2 + 2*x + 1
    
````

Use slicing to get the under and over sum, and the trapezoidal solution

````python
n = 100  
dx = 1/n
print(dx)
xx = np.linspace(0, 1, n+1)
# print(xx)

yy = f(xx)

#print([xx]:-1)
sum1 = yy[:-1].sum() * dx
sum2 = yy[1:].sum() * dx
sum3 = 0.5*(sum1 + sum2)
sum4 = (yy[1:-1].sum() + 0.5*(yy[0] + yy[-1])) * dx

print(sum1, sum2, sum3, sum4)
````

The midpoint method needs a little more work, but we will see it is already more accurate:

````python
xm = (xx[1:] + xx[:-1]) / 2
print(xm)

ym = f(xm)
sum5 = ym.sum() * dx
print(sum5) # QUESTION: How did you make this smaller?
````

Now use the advertised `quad` integrator from `scipy.integrate`

````python
sum, err = quad(f, 0, 1)
print(sum)
````

Integrate a function to infinity with `np.inf`. Pick a function we know can be integrated to a finite value

````python
def h(x):
    return np.exp(-x**2)

answer = np.sqrt(np.pi)/2
sum, err = quad(h, 0, np.inf)
print(sum, err, sum - answer)
````

Compute and plot the **indefinite integral** of the function.

$$
     f(x) = {{x} \over {\sqrt{1+x^2}}}       ~~~~\Leftrightarrow~~~~~     F(x) = \sqrt{1+x^2} - 1
$$
such that
$$
     f(0) = 0~, ~~~~  F(0) = 0
$$

````python
f1 = lambda x: x/np.sqrt(1 + x**2)  #f(x)
f2 = lambda x: np.sqrt(1 + x**2) -1 #F(x)

def F1(x):
    return quad(f1, 0, x)[0]

nx = 101
xmax = 4
x = np.linspace(0, xmax, nx)

f = f1(x)
F = np.zeros(nx)
d = np.zeros(nx)

for i in range(nx):
    F[i] = F1(x[i])
    d[i] = F[i] - f2(x[i])

print(d.mean(), d.std())
print(F[0], F[-1])
````

Plot $f(x)$ and $F(x)$ with two different axes, using `ax1.twinx()` to make a second axis on the right

Add additional parameters to the user defined function.

Specify them with the args tuple in the call to `quad()`.

````python
def gauss0(x, amp = 1, sigma = 1):
    return amp*np.exp(-x**2/(2**sigma**2))

sum1, err1 = quad(gauss0, 0 , 1)
print(sum1, err1)

sum2, err2 = quad(gauss0, 0 , 1, args = (2,5))
print(sum2, err2)
````
