1. Generating two dimensional arrays comes in many flavors, and with some warnings.
Note where the columns and rows go, and note the data type in this matrix `a`:
```
nx = 5                                     # number of COLUMNS
ny = 4                                     # number of ROWS
a = np.zeros(nx*ny).reshape(ny,nx)         # ’a’ is a float
a = np.zeros((ny,nx))                      # no reshape needed
```

### Low level python writing data

Starting simple, we want to write a small 3 x 4 table as follows (you could use an editor)
```
 1   2   3   4
11  12  13  14
21  22  23  24
```
so in the value the decimal gives the **row**, the single unit the **column**. 

We can use a multiline triple-string, and with low level python write the string to a file using **open/write/close**

This could be useful if you copy/paste something from another source.

````python
text = """ # a small table
1   2   3   4
11  12  13  14
21  22  23  24
"""
# print(text)

f = open('small.tab' , 'w')
f.write(text)
f.close()
print("wrote small.tab")
````

Inspect the file with `%more` or via the file explorer in the left pane . Use **control-B** to toggle the file explorer in view.

Turns out `more` and `!more` also work in JLD

````python
%more small.tab
````

````python
!more small.tab
````

now let's do this programmatically

````python
nx = 3 #number of columns
ny = 4 # number or rows

a = np.arange(nx*ny).reshape(ny,nx)
a = np.zeros((ny,nx))

for row in range(ny):
    for col in range(nx):
        a[row,col] = 10*row + col

a[0,0] = 0.45
print(a)
print(a.shape)
print(type(a[0,0]))
````

Notice the difference between python starting at 0, and (most?) humans starting at 1.

````python
np.savetxt('a.tab' , a, fmt = "%5.2f") #integer %d or float %g
np.savetxt('a.tab' , a, fmt = "%d", delimiter = ',')
%more a.tab
````

### Reading data

````python
b = np.loadtxt("a.tab" , delimiter = ',') #dtype=int #QUESTION: Purpose of delimiter?
print(b)
print(b.shape)
print(type(b[0,0]))
````

````python
print(b[1])   #print second row
print(b[:,1]) #print second column
````

### Stacking Columns

Create two arrays containing random numbers with `np.random.random`

Also create a linear time array.

Write the arrays to a text file by packing them into a tuple.
Use `np.transpose` to make the data column-wise if desired.

````python
seed = 123
np.random.seed(seed)
d1 = np.random.random(ny)
d2 = np.random.random(ny)
t1 = np.arange(ny)
# print(t1,d1,d2)

table = np.column_stack((t1,d1,d2))
print(table)
np.savetxt('random.tab' , table,  fmt = "%g" , header = "t, r1, r2")
%more random.tab
````

````python
data = np.loadtxt('random.tab').T
print(data)
print(data[0])
````

````python
fig, ax = plt.subplots()
ax.plot(data[0], data[1], 'ro-', label = "random 1")
ax.plot(data[0], data[2], 'go-', label = "random 2")
````
