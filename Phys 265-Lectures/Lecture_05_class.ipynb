## Simple pyPlot (style 1)

Plot the function with pyplot. 

````python
def f(x):
    return x*x
def g(x): 
    return x+x 
    
x = np.linspace(-2,2,11) 
print(x) 

y = f(x)
z = f(y)
t = np.tanh(x)
print(f'y={y}\n z = {z}\n t={t}')
````

````python
plt.plot(x,y)
plt.xlabel('X axis [x-unit]')
plt.ylabel('Y axis [y-unit]')
plt.title('My simple plot');
plt.suptitle('MATLAB lookalike');
````

### Object-Oriented (OO) matplotlib (style 2)

###  Style 2a:  figure and axis

````python
fig, ax1 = plt.subplots()
ax1.plot(x,y)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('My title for 00 plot')
fig.suptitle('00 plot')
fig.tight_layout() #make it nicer
````

### Style 2b: figure and multiple axis

````python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2) 
ax1.plot(x,y)
ax2.plot(y,x)
ax1.set_title('x vs y')
ax2.set_title('y vs x')

ax1.set_ylabel('y')
ax2.set_ylabel('x')

fig.tight_layout()
````

### Two types of plot

In the homeworks we sometimes ask for the **simple** plot, quick and dirty. And other times (labs especially) a **nice** plot is needed.  

1. A **simple** plot shows data or a function over some range of x. Three or four python statements are often sufficient to make a simple plot.
2. A **nice** plot includes embellishments that may require 10 to 15 python statements. Here is some guidance:
   1. Label the axes, with units specified in parentheses. Particularly the x-axis.
   2. The axis label font size should be increased (the default size is too small).
   3. The axis limits should be explicitly specified (do not rely on pyplot’s default choices).
   4. If there is more than one function shown, then a legend is required.
   5. The legend font size needs to be increased.
   6. If there is more than one function shown, then the color, markerstyle, and/or linestyle of each should be chosen to achieve clarity and readability.
   7. The plot should have a title, possibly with a larger fontsize.
   8. Conclude the plotting code with
```
       plt.tight_layout()
       plt.savefig(’myfigure.pdf’)         # or 'png'
```
