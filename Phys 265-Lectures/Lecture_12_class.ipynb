## Random Number Generators

We are often interested in generating *random numbers*.
For example, when picking raffle winners.

Generating *truly random* numbers is very hard.
It relies on access to a physical system which has genuine, random uncertainty -- e.g. throwing a pair of dice.
But throwing dice is slow! And not reproducible.

In practice, we use *pseudorandom* number generators (PRNGs).
These are *algorithms* which generate a *seemingly* random series of numbers.
* The sequence is in fact pre-determined, and controlled by a *seed* (or seeds).
* That is, the N-th number depends on the N-1-th number, is some pre-determined way.
* If you know the seed, you can reproduce the entire series.

People work hard to design methods that are good, e.g. have low periodicity, weak correlations/patterns between numbers, and so on.

Comment: as a general rule: "don't roll your own crypto."
Cryptography is *hard*. Use someone else's algorithm!

### Numpy for RNGs

Numpy contains RNG routines. In fact, when you call `import numpy as np`, a seed gets created behind the scenes! What numpy calls the "global random seed." The easiest routine is the one we used last class: `np.random.random`.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(12)
a = np.random.rand(4)
print(a)
```

But relying on the global random seed is poor practice!
* What if some other code you import tries to reset it, or change it!?
* Or what if some other routine uses the RNG?

Then you don't know how many calls were actually made, and reproducing the sequence doesn't work.

The modern, best-practice way is to create a RNG object via the `np.random.default_rng` method. And then use it. If a sub-routine needs to generate random numbers, pass the object.

The default RNG in Numpy is the PCG64 algorithm -- an *excellent* algorithm (for several reasons) published in 2014 by Prof. Melissa O'Neill of Harvey Mudd College.

```python
seed = 12
rng = np.random.default_rng(seed)
print(rng)
a = rng.random(4)
print(a)
```

Numpy also contains modules for generating cryptographically strong RNGs. This is via the `secrets` module. 

There is also some subtelty to managing RNGs in the context of parallel processing, where tools like `SeedSequence` are helpful.

But we won't cover them in this class. Take 465!

### Random Distributions

By default, numpy returns a random number that is *uniformly* distributed from [0, 1) -- that is, including zero, but not including 1.

You can also ask numpy to draw the numbers randomly according to a distribution, or in a certain way:
* integers
* gaussian distributions
* And many more... https://numpy.org/doc/stable/reference/random/generator.html#distributions

```python
five_integers = rng.integers(low = 5, high = 10, size = 5)
print(five_integers)

seven_guassian = rng.normal(loc = 0, scale = 1, size = 7)
print(seven_guassian)
```

## Histograms

Histograms are ways of visualizing data *frequency*. Simplest case is 1D.

The output of the histogram call is the:
* Counts per bin
* The bin *edges*
* "patches": the "artist" (a matplotlib concept) used to create the histogram.

Matplotlib gives us many options. We can...
* Control the binning
* Normalize the histogram (turn into a *density* plot)
* Control visualization details -- e.g., filled, outline, etc.

```python
#generate 0,000 random numbers, uniformly distributed from -10 to 10
data_uniform = rng.uniform(-10, 10, 10000)
data_gaussian = rng.normal(0, 2, 10000)

fig, ax = plt.subplots(1, 1)
# ax.hist(data_uniform)
my_bins = np.linspace(-10, 10, 100)
ax.hist(data_uniform, bins = my_bins, density = True)
counts, bins, patchs = ax.hist(data_gaussian, bins = my_bins, density = True, histtype = "step")
ax.set_yscale("log")
```

### 2D Histograms

Can also visual numbers in 2D. When you histogram in 2D, you get returned back:
* counts (events per bin)
* xedges, yedges: the bin edges in x and y
* image: the `QuadMesh` mappable object (2D rectangular grid for plotting)

```python
x_vals = rng.normal(0, 1, size = int(1E5))
y_vals = rng.normal(0, 1, size = int(1E5))

fig, ax = plt.subplots(1, 1)
counts, xedges, yedges, im = ax.hist2d(x_vals, y_vals, bins = 50, cmin = 1) #QUESTION: Different from lecture and cmin
ax.set_aspect("equal")
cbar = fig.colorbar(im, ax = ax)
```

### A comment on colorbars

The matplotlib default colormap -- viridis -- should be used with care.
* It is a very clever colormap, but, it is very hard to tell the difference between "1" (something) and "0" (nothing).
* I advise setting `cmin=1` if you want to use it.

If you choose a different colorbar, choose carefully! The `jet` ("rainbow") colormap is usually a particularly bad choice, for reasons you can read about online: https://www.domestic-engineering.com/drafts/viridis/viridis.html

## Plots with Errorbars

In physics, we always strive to know our errors! Matplotlib can make plots with errorbars.

Throw an object straight up from initial height of $z_0 = 500$m and velocity $v_0 = 40$ m/s. Measure height vs time, and measuring tool has uncertainties:
* 0.5 seconds on every time point
* 3% on the height

```python
g = -9.8 #m/s2
z0 = 500 # m
v0 = 49 #m/s
t = np.linspace(0, 10, 11)
z = z0 + v0*t + 0.5*g*t**2

t_err = np.full_like(t, 0.5)
z_err = z*0.03

fig, ax = plt.subplots(1,1)
ax.errorbar(t, z, xerr = t_err, yerr = z_err, fmt = "o")
```
